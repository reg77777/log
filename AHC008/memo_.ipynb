### **休憩**



以下はスコア計算を並列処理するコード

テスターのフォルダに解法コード(slove.cpp)と一緒に入れて実行している

```python
import subprocess
from tqdm import tqdm

pl=[]
l=[]
N=1000
from subprocess import PIPE
for i in range(N):
    l.append(i)
M=12
for i in tqdm(l):
    s=str(i).zfill(4)
    out=open(f'./out/{s}.out','w')
    p=subprocess.Popen([f'cargo run --release --bin tester ./a.out < in/{s}.txt'],shell=True,stdout=out,stderr=out)
    pl.append(p)
    if (i+1)%M==0 or (i+1)==N:
        for p in pl:
            p.wait()
        pl=[]

scores=[]
for i in l:
    s=str(i).zfill(4)
    f=open(f'./out/{s}.out','r')
    txts=f.readlines()
    f.close()
    score=int(txts[len(txts)-1].strip().replace('Score = ',''))
    scores.append(score)

import numpy as np
print(np.mean(scores))
```
